1. Importing Required Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

2. Loading the Netflix Dataset and Displaying the dataset

```python
df = pd.read_csv('netflix_titles.csv')
df.head()
```

4. Understanding Dataset Information

```python
df.info()
```

5. Checking Missing Values

```python
df.isnull().sum()
```

```python
print (df.columns)
```

6. Data Cleaning and Handling Missing Values

```python
df['director'] = df['director'].fillna('Unknown')

df['cast'] = df['cast'].fillna('Not Available')

df['country'] = df['country'].fillna('Unknown')

df['rating'] = df['rating'].fillna(df['rating'].mode()[0])

df = df.dropna(subset=['date_added'])
```

```python
df.isnull().sum()
```

7. Removing Duplicate Records

```python
df.drop_duplicates(inplace=True)
```

```python
df.duplicated().sum()
```

8. Converting Date Columns

```python
df['date_added'] = pd.to_datetime(df['date_added'], errors='coerce')
```

```python
df['date_added'].head()
```

9. Data Visualization

9.1. Movies vs TV Shows Analysis

```python
df['type'].value_counts().plot(kind='bar')

plt.title('Movies vs TV Shows')

plt.xlabel('Type')

plt.ylabel('Count')

plt.show()
```

9.2. Top 10 Countries Analysis

```python
df['country'].value_counts().head(10).plot(kind='bar')

plt.title('Top 10 Countries')

plt.xlabel('Country')

plt.ylabel('Count')

plt.show()
```

9.3. Ratings Distribution Analysis

```python
df['rating'].value_counts().plot(kind='pie', autopct='%1.1f%%')

plt.title('Ratings Distribution')

plt.show()
```

9.4. Release Year Analysis

```python
df['release_year'].value_counts().head(15).plot(kind='bar')

plt.title('Top Release Years')

plt.xlabel('Year')

plt.ylabel('Count')

plt.show()
```
